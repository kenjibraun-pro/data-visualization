```
import numpy as np
import matplotlib.pyplot as plt

x = np.linspace(0, 30, num=150)
y_1 = 3/4*np.sin(x)
y_2 = 1/4*np.sin(x-np.pi/3)
y = y_1+y_2

plt.figure(figsize=(10,5))
plt.plot(x, y_1, 'b--', label='3/4*sin(x)')
plt.plot(x, y_2, 'y--', label='1/4*sin(x-pi/3)')
plt.plot(x, y, 'r', label='3/4*sin(x)+1/4*sin(x-pi/3)')
plt.legend(loc='upper right')
plt.ylim((-1.2, 1.5))
plt.title('Two sin wave with a different phase')
plt.grid()
plt.show()
```

```
y_1 = 3/4*np.sin(x)
y_2 = 1/4*np.sin(3*x)
y = y_1+y_2

plt.figure(figsize=(10,5))
plt.plot(x, y_1, 'b--', label='3/4*sin(x)')
plt.plot(x, y_2, 'y--', label='1/4*sin(3*x)')
plt.plot(x, y, 'r', label='3/4*sin(x)+1/4*sin(3*x)')
plt.legend(loc='upper right')
plt.ylim((-1, 1.5))
plt.title('Two sin wave with a different frequency')
plt.grid()
plt.show()
```

```
y_1 = 4/np.pi*np.sin(x)
y_2 = (4/(3*np.pi))*np.sin(3*x)
y_3 = (4/(5*np.pi))*np.sin(5*x)
y = y_1+y_2+y_3

plt.figure(figsize=(15,5))
plt.plot(x, y_1, 'b--', label='y_1')
plt.plot(x, y_2, 'y--', label='y_2')
plt.plot(x, y_3, 'g--', label='y_3')
plt.plot(x, y, 'r', label='y')
plt.legend(loc='upper right')
#plt.ylim((-1, 1.5))
plt.title('Three sin wave with a different frequency')
plt.grid()
plt.show()
```

```
x = np.linspace(0, 30, num=300)
y = 0
for i in range(1,100,2):
  y += (4/(i*np.pi))*np.sin(i*x)   # 1, 3, 5, 7, ...

plt.figure(figsize=(15,5))
plt.plot(x, y, 'r', label='y')
plt.legend(loc='upper right')
#plt.ylim((-0.5, 0.5))
plt.title('Many sin wave with a different frequency')
plt.grid()
plt.show()
```

```
y = 0
for i in range(1,29):
  if i%2 == 1:
    y += (4/(i*np.pi))*np.sin(i*x)   # 1, 3, 5, 7, ...
  if i%2 == 0:
    y += (-4/(i*np.pi))*np.sin(i*x)

plt.figure(figsize=(15,5))
plt.plot(x, y, 'r', label='y')
plt.legend(loc='upper right')
#plt.ylim((-1, 1.5))
plt.title('Many sin wave with a different frequency')
plt.grid()
plt.show()
```
